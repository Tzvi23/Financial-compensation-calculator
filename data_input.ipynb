## Departure probabilities

```python
# Departure probabilities
def get_prob(age, param):
    # Dict: key: (age, age) : value: [dismissal, resignation]
    prob_table = {(18, 29): [0.07 , 0.2], 
     (30, 39): [0.05, 0.13], 
     (40, 49): [0.04, 0.1],
     (50, 59): [0.03, 0.07],
     (60, 67): [0.02, 0.03]
    }
    for key in prob_table.keys():
        if key[0] <= age <= key[1]:
            return prob_table[key]
        else:
            return None
```
